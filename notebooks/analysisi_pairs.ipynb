```python
#Subir de notebook/ a la carpeta raiz del proyecto
%cd ..
import src.core as core
```

```python


closes = core.load_and_prepare_closes()      # usa tus datos ya descargados
returns = core.compute_log_returns(closes)

base = closes / closes.iloc[0]
base.plot(logy = True)
returns.plot()
```

```python
#Imprimir correlacciones
corr = core.rolling_correlation_dual(returns, "BTC-USD", "QQQ", window_short=30, window_medium=90, window_long = 180)
ax = corr.plot(figsize=(10, 4))  # ax es un objeto Axes
ax.set_title("Correlacin de Pearson para los retornos logartmicos diarios de BTC-USD - QQQ")
ax.set_xlabel("Fecha")
ax.set_ylabel("Correlacin")
ax.grid(True)

```

```python
#Imprimir beta y spread de los retornos

log_prices = core.compute_log_prices(closes)
spread, beta = core.compute_spread(log_prices, "BTC-USD", "QQQ")

spread.plot(figsize=(10, 4))
ax = spread.plot(figsize=(10, 4))  # ax es un objeto Axes
ax.set_title(f"Spread BTC-USD - QQQ (beta = {beta:.4f})")
ax.set_xlabel("Fecha")
ax.set_ylabel("Spread")
ax.grid(True)


```
